```python
%matplotlib qt
import numpy as np
from pathlib import Path
from utils.utils import run
from utils.train import SVMclassifier
from lima_beans.bean_utils import QtLinePicker, process_and_save_objects
```

```python
# CELL1
# -- provide path to the reflectance numpy file

reflectance_np_file = 'path/to/reflectance/numpy/file/51_000_0008.npy'
```

```python
# CELL2
reflectanceX = np.load(reflectance_np_file)

filename_without_ext = Path(reflectance_np_file).stem
print(f'Processing file: {filename_without_ext}')
```

### Extract data and train SVM classfier

Data Extraction

```python
# CELL3

X, y = run(reflectanceX)
```

Train SVM Classifier

```python
# CELL4

svm_classifier = SVMclassifier()
svm_mask, trained_classifier = svm_classifier.train_svm(reflectanceX, X, y)

# visualize the mask and a random band (check for noise)
svm_classifier.display(hypercube=reflectanceX, mask1=svm_mask)
```

Postprocess

```python
# CELL5
# max_size=20 to tune 

final_mask = svm_classifier.post_process(mask=svm_mask, max_size=20)
svm_classifier.display(hypercube=reflectanceX, mask1=svm_mask, mask2=final_mask)
```

### Data preparation, processing, and saving 

```python
# CELL6

# 1. Open the window and click between the groups on Band 70
# click the area where upper and lower beans can be differentiated

line_picker = QtLinePicker(reflectanceX, display_band=70)
custom_split_y = line_picker.get_y()

print(f"Dividing line set at Y = {custom_split_y}")
```

```python
# CELL7

# 2. Run the processing function with custom split line
if custom_split_y is not None:
    
    process_and_save_objects(reflectanceX, final_mask, custom_split_y, 
                             output_dir=f'results/{filename_without_ext}')

else:
    print("No line was selected. Please run the line picker again.")
```
