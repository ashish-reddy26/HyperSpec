```python
# PLSR Pipeline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
from sklearn.model_selection import train_test_split, KFold, cross_val_predict
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import root_mean_squared_error, r2_score, mean_absolute_percentage_error
from scipy.stats import pearsonr

def run_plsr_pipeline(df, feature_cols, target_col, group_col, max_comp=20, test_size=0.2, random_state=1, window_length=9, polyorder=2, deriv=0):
    """
    Runs the PLSR modeling pipeline for a specified target variable.
    """
    print(f"==========================================")
    print(f"Modeling Target: {target_col.upper()}")
    print(f"==========================================")

    # 1. DEFINE DATA
    X = df[feature_cols]
    y = df[target_col] 
    groups = df[group_col] 

    # 2. STRATIFIED SPLIT
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, 
        test_size=test_size, 
        stratify=groups,
        random_state=random_state
    )

    # 3. SPECTRAL PREPROCESSING
    X_train_smooth = savgol_filter(X_train, window_length=window_length, polyorder=polyorder, axis=1, deriv=deriv)
    X_test_smooth = savgol_filter(X_test, window_length=window_length, polyorder=polyorder, axis=1, deriv=deriv)

    # 4. OPTIMIZE n_components VIA CROSS-VALIDATION
    mse_cv_scores = []
    kf = KFold(n_splits=10, shuffle=True, random_state=random_state)

    for i in range(1, max_comp + 1):
        pls_cv = PLSRegression(n_components=i)
        y_cv_pred = cross_val_predict(pls_cv, X_train_smooth, y_train, cv=kf)
        rmse_cv = root_mean_squared_error(y_train, y_cv_pred)
        mse_cv_scores.append(rmse_cv)

    optimal_components = np.argmin(mse_cv_scores) + 1 
    print(f"Optimal n_components found: {optimal_components}")

    # Plot the cross-validation curve
    plt.figure(figsize=(8, 5))
    plt.plot(range(1, max_comp + 1), mse_cv_scores, marker='o', linestyle='--')
    plt.xlabel('Number of PLS Components')
    plt.ylabel('RMSE (Cross-Validation)')
    plt.title(f'Determining Optimal n_components for {target_col}')
    plt.xticks(range(1, max_comp + 1))
    plt.axvline(x=optimal_components, color='r', linestyle=':', label=f'Optimal ({optimal_components})')
    plt.legend()
    plt.show()

    # 5. TRAIN FINAL MODEL WITH OPTIMAL COMPONENTS
    pls_final = PLSRegression(n_components=optimal_components)
    pls_final.fit(X_train_smooth, y_train)

    # 6. EVALUATE FINAL MODEL
    # Train Metrics
    y_train_pred = pls_final.predict(X_train_smooth)
    rmse_train = root_mean_squared_error(y_train, y_train_pred)
    r2score_train = r2_score(y_train, y_train_pred)
    
    # Test Metrics
    y_test_pred = pls_final.predict(X_test_smooth)
    if y_test_pred.ndim > 1:
        y_test_pred = y_test_pred.squeeze()

    mape = mean_absolute_percentage_error(y_test, y_test_pred)
    rmse_test = root_mean_squared_error(y_test, y_test_pred)
    r2score_test = r2_score(y_test, y_test_pred)
    r_value, p_value = pearsonr(y_test, y_test_pred)

    print(f'\n--- Final Model Performance ---')
    print(f'Train RMSE: {rmse_train:.4f}, Train R2 Score: {r2score_train:.4f}')
    print(f'Test RMSE:  {rmse_test:.4f}, Test R2 Score:  {r2score_test:.4f}')
    print(f'Test MAPE:  {mape:.4f}')
    print(f'Pearson R:  {r_value:.4f}, P-value: {p_value:.4e}\n')

    # Store results in a dictionary to return
    results = {
        'optimal_components': optimal_components,
        'train_rmse': rmse_train,
        'train_r2': r2score_train,
        'test_rmse': rmse_test,
        'test_r2': r2score_test,
        'test_mape': mape,
        'pearson_r': r_value,
        'p_value': p_value
    }

    return pls_final, results
```

```python
# Define features and group columns
feature_columns = df.columns[2:203] 
group_column = 'Group'

# List all the target variables you want to predict
target_variables = [
    'surface_area', 
    'pore_size', 
    'pore_volume', 
    'water_holding_capacity', 
    'carbon_content', 
    'density'
]

# Dictionary to store the models and metrics for each target
all_results = {}

# Loop through each target variable and run the pipeline
for target in target_variables:
    # Ensure the target column exists in the dataframe before running
    if target in df.columns:
        model, metrics = run_plsr_pipeline(
            df=df, 
            feature_cols=feature_columns, 
            target_col=target, 
            group_col=group_column,
            max_comp=20
        )
        # Store the results
        all_results[target] = {'model': model, 'metrics': metrics}
    else:
        print(f"Warning: Column '{target}' not found in dataframe.")

# Example: Accessing the test R2 score for carbon_content after the loop finishes
# print(all_results['carbon_content']['metrics']['test_r2'])
```
